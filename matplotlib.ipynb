```python
heart_disease = pd.read_csv('013 heart-disease.csv')
heart_disease
```

```python
heart_disease['age'].plot.hist(bins=80);
```

```python
heart_disease.plot.hist(figsize=(10,10), subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
ax.set(title='Heart Disease and Age');
#ax.set_xlim([45, 100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart Disease and Age',
       xlabel='Age',
       ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(), title='Target');

ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=1,
                       figsize=(10,10),
                       sharex=True)

scatter = ax[0].scatter(x=over_50['age'],
                        y=over_50['chol'],
                        c=over_50['target'])
ax[0].set(title='Heart Disease and Age',
          ylabel='Cholesterol')

ax[0].legend(*scatter.legend_elements(), title='Target')

ax[0].axhline(over_50['chol'].mean(),
              linestyle='--');

scatter = ax[1].scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'])
ax[1].set(title='Heart Disease and Age',
          xlabel='Age',
          ylabel='Max Heart Rate')
ax[1].legend(*scatter.legend_elements(), title='Target')
ax[1].axhline(over_50['thalach'].mean(),
              linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

ax.set(title='Heart Disease and Age',
       xlabel='Age',
       ylabel='Cholesterol')
ax.legend(*scatter.legend_elements(), title='Target');

ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=1,
                       figsize=(10,10),
                       sharex=True)

scatter = ax[0].scatter(x=over_50['age'],
                        y=over_50['chol'],
                        c=over_50['target'],
                        cmap='winter')
ax[0].set(title='Heart Disease and Age',
          ylabel='Cholesterol')
ax[0].set_xlim([50,80])
ax[0].legend(*scatter.legend_elements(), title='Target')

ax[0].axhline(over_50['chol'].mean(),
              linestyle='--');

scatter = ax[1].scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'],
                        cmap='winter')
ax[1].set(title='Heart Disease and Age',
          xlabel='Age',
          ylabel='Max Heart Rate')
ax[1].set_xlim([50,80])
ax[1].set_ylim([60,200])
ax[1].legend(*scatter.legend_elements(), title='Target')
ax[1].axhline(over_50['thalach'].mean(),
              linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
